```python
df = pd.read_csv('QueryResults.csv', names=['DATE', 'TAG', 'POSTS'], header=0)
```

```python
df.groupby('TAG').sum()
```

```python
df.groupby('TAG').count()
```

```python
df.DATE = pd.to_datetime(df.DATE)
df.head()
```

```python
df.head(20)
```

```python
reshaped_df = df.pivot(index='DATE', columns='TAG', values='POSTS')
```

```python
reshaped_df.shape
```

```python
reshaped_df.fillna(0, inplace=True) 
```

```python
reshaped_df.head(50)
```

```python
reshaped_df.isna().values.any()
```

```python
plt.figure(figsize=(16,14))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0, 38000)
plt.plot(reshaped_df.index, reshaped_df.javascript)
plt.plot(reshaped_df.index, reshaped_df.python)
plt.plot(reshaped_df.index, reshaped_df.java)
plt.plot(reshaped_df.index, reshaped_df.swift)
plt.plot(reshaped_df.index, reshaped_df['c#'])
plt.plot(reshaped_df.index, reshaped_df.c)
plt.plot(reshaped_df.index, reshaped_df['c++'])
plt.plot(reshaped_df.index, reshaped_df.ruby)
plt.plot(reshaped_df.index, reshaped_df.r)
plt.plot(reshaped_df.index, reshaped_df.php)
plt.legend(['javascript', 'python', 'java', 'swift', 'c#', 'c', 'c++', 'ruby', 'r', 'php'])
plt.xlabel('Years', fontsize=18, color='coral')
plt.ylabel('Number of Stack Overflow Posts per year', fontsize=18, color='coral')
plt.title('Programming Language Trend on Stack Overflow', fontsize=30, color='maroon')
```

```python
reshaped_df = reshaped_df.rolling(window=3).mean()
plt.figure(figsize=(16,14))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0, 38000)
plt.plot(reshaped_df.index, reshaped_df.javascript)
plt.plot(reshaped_df.index, reshaped_df.python)
plt.plot(reshaped_df.index, reshaped_df.java)
plt.plot(reshaped_df.index, reshaped_df.swift)
plt.plot(reshaped_df.index, reshaped_df['c#'])
plt.plot(reshaped_df.index, reshaped_df.c)
plt.plot(reshaped_df.index, reshaped_df['c++'])
plt.plot(reshaped_df.index, reshaped_df.ruby)
plt.plot(reshaped_df.index, reshaped_df.r)
plt.plot(reshaped_df.index, reshaped_df.php)
plt.legend(['javascript', 'python', 'java', 'swift', 'c#', 'c', 'c++', 'ruby', 'r', 'php'])
plt.xlabel('Years', fontsize=18, color='coral')
plt.ylabel('Number of Stack Overflow Posts per year', fontsize=18, color='coral')
plt.title('Programming Language Trend on Stack Overflow', fontsize=30, color='maroon')
```
